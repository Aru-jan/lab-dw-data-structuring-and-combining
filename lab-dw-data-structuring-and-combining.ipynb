```python
import numpy as np
import pandas as pd
```

```python
url_1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
df_1 = pd.read_csv(url_1)
df_1.head(3)
```

```python
url_2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
df_2 = pd.read_csv(url_2)
df_2.head(3)
```

```python
url_3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"
df_3 = pd.read_csv(url_3)
df_3.head(3)
```

```python
df = pd.concat([df_1, df_2, df_3], axis=0, join="outer")
df.head(3)
```

```python
df["State_combined"] = df["ST"].combine_first(df["State"])
df.head(3)
```

```python
df = df.drop(columns=["ST", "State"])
df.head(3)
```

```python
df= df.rename(columns={"State_combined": "State"})
df.head(3)
```

```python
df["Gender_combined"] = df["GENDER"].combine_first(df["Gender"])
df.head(3)
```

```python
df = df.drop(columns=["GENDER", "Gender"])
df.head(3)
```

```python
df = df.rename(columns={"Gender_combined": "Gender"})
df.head(3)
```

```python
df.columns.tolist()
```

```python
new_order = ['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value', 'Income', 'Monthly Premium Auto', 
             'Number of Open Complaints', 'Policy Type', 'Vehicle Class', 'Total Claim Amount']
```

```python
df = df[new_order]
df.head(3)
```

```python
df.columns = [col.replace(" ","_").lower()for col in df.columns]
df.head(3)
```

```python
df["state"].value_counts()
```

```python
df["state"].isna().sum()
```

```python
df["state"] = df["state"].replace(r'.*AZ.*', 'Arizona', regex=True)
df["state"] = df["state"].replace(r'.*WA.*', 'Washington', regex=True)
df["state"] = df["state"].replace(r'.*Cali.*', 'California', regex=True)
```

```python
df["state"].value_counts()
```

```python
df["state"].isna().sum()
```

```python
df["gender"].unique().tolist()
```

```python
df["gender"] = df["gender"].replace(r'.*Femal.*', 'F', regex=True)
df["gender"] = df["gender"].replace(r'.*female.*', 'F', regex=True)
df["gender"] = df["gender"].replace(r'.*Male.*', 'M', regex=True)
```

```python
df["education"].unique().tolist()
```

```python
df["education"] = df["education"].replace(r'.*Bachelors.*', 'Bachelor', regex=True)
```

```python
df["education"].unique().tolist()
```

```python
df["customer_lifetime_value"] = df["customer_lifetime_value"].replace(r"%","", regex = True)
```

```python
df["vehicle_class"] = df["vehicle_class"].str.replace("Sports Car", "Luxury")
df["vehicle_class"] = df["vehicle_class"].str.replace("Luxury SUV", "Luxury")
df["vehicle_class"] = df["vehicle_class"].str.replace("Luxury Car", "Luxury")
```

```python
df.dtypes
```

```python
df["customer_lifetime_value"] = pd.to_numeric(df["customer_lifetime_value"])
df.dtypes
```

```python
complaints_split = df["number_of_open_complaints"].str.split("/", expand=False)
complaints_keep1 = complaints_split.str[1]
display(complaints_keep1)
df["number_of_open_complaints"] = complaints_keep1
```

```python
df["number_of_open_complaints"] = pd.to_numeric(df["number_of_open_complaints"])
df.dtypes
```

```python
df.tail(5)
```

```python
df.isna().sum()
```

```python
new_null = df.isna().sum(axis=1)
new_null
```

```python
new_null.unique()
```

```python
new_null.value_counts() 
```

```python
#since there are 2937 null values in 11columns, i will drop them
df.dropna(thresh = 3, inplace=True) 
df
```

```python
df.duplicated().sum()
```

```python
df= df.drop_duplicates()
```
